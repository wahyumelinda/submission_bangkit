## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
import matplotlib.pyplot as plt

sales_per_product = all_data.groupby('product_category_name')['price'].count().sort_values(ascending=False)

# Visualisasi produk dengan total penjualan tertinggi
plt.figure(figsize=(10, 6))
sales_per_product.head(10).plot(kind='bar', color='pink')
plt.title('Top 10 Produk dengan Total Penjualan Tertinggi')
plt.xlabel('Product Name')
plt.ylabel('Total Unit yang Terjual')
plt.xticks(rotation=45)
plt.show()
```

### Pertanyaan 2:

```python
import matplotlib.pyplot as plt

sales_per_city = all_data.groupby('customer_city')['customer_id'].count().sort_values(ascending=False)

# Visualisasi kota dengan penjualan terbanyak (Top 10)
plt.figure(figsize=(10, 6))
sales_per_city.head(10).plot(kind='bar', color='pink')
plt.title('Top 10 Kota dengan Total Penjualan Terbanyak')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

### Pertanyaan 3:

```python
sales_per_product = all_data.groupby('product_id').agg(
    total_sales=('order_id', 'count'),
    avg_price=('price', 'mean')
)

plt.figure(figsize=(8, 6))
plt.scatter(sales_per_product['avg_price'], sales_per_product['total_sales'], color='pink')
plt.title('Hubungan Harga Produk dengan Total Penjualan')
plt.xlabel('Harga Rata-rata Produk')
plt.ylabel('Total Units Sold')
plt.grid(True)
plt.show()
```

**Insight:**

- kota sao paulo merupakan kota dengan total penjualan tertinggi
- dari visualisasi yang didapatkan hubungan harga produk dengan total penjualan memiliki nilai -1 yang berarti semakin murah atau rendah harganya maka semakin tinggi penjualannya dan sebaliknya jika produk memiliki harga yang mahal maka penjualannya cenderung sedikit

## Analisis Lanjutan (Opsional)

```python
rfm_df = all_data.groupby(by="product_category_name", as_index=False).agg({
    "order_purchase_timestamp": "max",
    "order_id": "nunique", 
    "price": "sum"  
})

rfm_df.columns = ["product_category_name", "max_order_timestamp", "frequency", "monetary"]

rfm_df["max_order_timestamp"] = pd.to_datetime(rfm_df["max_order_timestamp"]).dt.date
recent_date = pd.to_datetime(all_data["order_purchase_timestamp"]).dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_timestamp", axis=1, inplace=True)

print(rfm_df.head())
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

colors = ["#FF69B4", "#FF69B4", "#FF69B4", "#FF69B4", "#FF69B4"]

# Grafik Recency
sns.barplot(y="recency", x="product_category_name", hue="product_category_name", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0], legend=False)
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis='x', labelsize=9)

# Grafik Frequency
sns.barplot(y="frequency", x="product_category_name", hue="product_category_name", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1], legend=False)
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=9)

# Grafik Monetary
sns.barplot(y="monetary", x="product_category_name", hue="product_category_name", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2], legend=False)
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=9)

plt.suptitle("Best Product Based on RFM Parameters (product_category_name)", fontsize=20)
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Mengatur layout agar tidak terpotong
plt.show()
```
